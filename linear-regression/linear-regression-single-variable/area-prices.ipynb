```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("area-prices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area (sqft)')
plt.ylabel('price (INR)')
plt.scatter(df.area, df.price, color = 'red', marker = '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[5500]])
```

```python
reg.coef_ # shows the value of the slope (m) in equation y = mx + b
```

```python
reg.intercept_ # shows the value of the intercept (b) in equation y = mx + b
```

```python
# y = mx + b
2000*5500+1.862645149230957e-09
```

```python
%matplotlib inline
plt.xlabel('area (sqft)')
plt.ylabel('price (INR)')
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.plot(df.area, reg.predict(df[['area']]), color = 'blue')
```
